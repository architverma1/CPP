```python
flat = np.hstack((np.reshape(fit['mu'], (24,2)), np.reshape(fit['a'], (24,4)), np.reshape(fit['b'], (24,4))))
```

```python
param_names = ['$\mu_{KTR}$', '$\mu_{GPF}$','$a_{KTR}$', '$a_{KTR > GFP}$', '$a_{GFP > KTR}$',
              '$a_{GFP}$', '$b_{KTR}$', '$b_{KTR > GFP}$', '$b_{GFP > KTR}$',
              '$b_{GFP}$']
```

```python
flat_norm = np.divide(flat - np.mean(flat, axis = 0), np.std(flat, axis = 0))
```

```python
fig, ax = plt.subplots()
im = ax.imshow(flat_norm, cmap = 'viridis')
ax.set_xticks(np.arange(len(param_names)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(param_names)
ax.set_yticklabels(labels)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
fig.tight_layout()
cbar = ax.figure.colorbar(im, ax=ax)
plt.show()
```
